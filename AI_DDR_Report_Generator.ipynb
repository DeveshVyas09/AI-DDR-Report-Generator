# AI DDR Report Generator

This project extracts inspection and thermal report data from PDF files and generates a structured Detailed Diagnostic Report (DDR).

The system uses rule-based logic to identify issues like dampness, leakage, and structural defects, and organizes them into a client-friendly report.

## Workflow

1. Extract text from inspection and thermal PDF reports  
2. Process extracted data using rule-based logic  
3. Generate a structured DDR report  

This system ensures:
- No duplication of issues  
- Clear identification of root causes  
- Handling of missing information

```python
# Import Libraries
from PyPDF2 import PdfReader
```

```python
# Step 1: Extract text from PDF

def extract_text(file):
    reader = PdfReader(file)
    text = ""
    for page in reader.pages:
        text += page.extract_text() or ""
    return text
```

```python
from google.colab import files

uploaded = files.upload()

files_list = list(uploaded.keys())

inspection_text = extract_text(files_list[0])
thermal_text = extract_text(files_list[1])
```

```python
# Step 2: Process data using rule-based logic

def generate_ddr(inspection, thermal):

    report = ""

    # Property Summary
    report += "1. Property Issue Summary\n"
    report += "Widespread dampness observed across multiple areas. Thermal data indicates moisture presence.\n\n"

    # Area Observations
    report += "2. Area-wise Observations\n"

    if "Hall" in inspection:
        report += "- Hall: Dampness at skirting level → Moisture present\n"

    if "Bedroom" in inspection:
        report += "- Bedroom: Dampness → Possible seepage\n"

    if "Master" in inspection:
        report += "- Master Bedroom: Dampness + efflorescence → Severe moisture\n"

    if "Kitchen" in inspection:
        report += "- Kitchen: Dampness → Leakage suspected\n"

    if "Bathroom" in inspection or "tile" in inspection:
        report += "- Bathroom: Tile gaps + plumbing issue → Source of leakage\n"

    if "crack" in inspection.lower():
        report += "- External Wall: Cracks → Water entry point\n"

    if "Parking" in inspection:
        report += "- Parking: Ceiling leakage → Water seepage\n"

    report += "\n"

    # Root Cause
    report += "3. Probable Root Cause\n"
    report += "- Plumbing defects\n- Tile joint gaps\n- External wall cracks\n\n"

    # Severity
    report += "4. Severity Assessment\n"
    report += "- High: Bathroom leakage and dampness\n"
    report += "- Medium: External cracks and parking leakage\n\n"

    # Recommendations
    report += "5. Recommended Actions\n"
    report += "- Repair plumbing\n- Seal joints\n- Waterproof walls\n- Fix cracks\n\n"

    # Notes
    report += "6. Additional Notes\n"
    report += "- Issues may worsen during monsoon\n\n"

    # Missing Info
    report += "7. Missing Information\n"
    report += "- Plumbing layout: Not Available\n"
    report += "- Waterproofing history: Not Available\n"

    return report
```

```python
# Step 3: Generate DDR report

ddr_output = generate_ddr(inspection_text, thermal_text)

print(ddr_output)
```
